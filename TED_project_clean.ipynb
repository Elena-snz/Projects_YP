**Тема:** исследование истории TED-конференций

**Описание:** С 1984 года в Калифорнии (США) проходит конференция, собирающая на одной сцене ученых, политиков, предпринимателей, художников, музыкантов, инженеров и многих других. Она называется TED — Technology, Entertainment, Design (Наука, Искусство, Культура). Слоган конференции - «Идеи, достойные распространения», объясняет ее главную задачу — рассказать об интересных идеях как можно большему количеству людей.

**Данные:** дано три файла, которые хранят данные о выступлениях, четвертый файл - справочник конференций, пятый файл - справочник авторов выступления. Предобработки данных не проводилось.

**Исследование:** будут представлены визуализации данных по следующим темам:

    - история выступлений
    - тематики выступлений
    - авторы выступлений
    - число просмотров выступлений
    
**Выводы:** 

* Чаще всего выступления на конференции TED проходили в США, Канаде и Великобритании
* ТОП-5 тематик, по которым проводились выступления: наука,технологии, культура, общество, глобальные проблемы
* По медианному числу просмотров популярны совсем другие категории: философия, ментальное здоровье, принятие решений, счастье и дети.
* После 2020 года на первый план вышли такие тематики выступлений, как технологии, общество и глобальные проблемы. А вот науку и культуру обсуждать перестали. Думаю, это связано с тем, что общество столкнулось с эпидемией Covid-19 и мир оказался не готов к ней.
* Выступления по тематике entertainment чаще вызывают смех и аплодисменты. 
* Выступления длительностью не более 20 минут набирают больше просмотров, чем более длинные выступления. 
* Максимальное число просмотров - 72 млн. 
* Самое длинное выступление 62.5 минуты.
* Преобладающие области деятельности авторов - писатель, артист, дизайнер, журналист
* Чаще всего авторы выступают по одному разу. 
* Медианное число просмотров конференций TED практически не меняется, что говорит о популярности такого формата лекций и тем, которые там поднимаются. Самые активные по числу просмотров годы 2004, 2013, 2015 и 2019. 
* После 2019 года число просмотров пошло на спад. Но делать выводы о потере интереса зрителей пока рано, нужна дополнительная информация за последующие годы.
